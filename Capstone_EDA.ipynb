# Replacing Data

```python
a1_loop = {'None': 0,
           'Norm': 1,
           '>7': 2,
           '>8': 3}

serum_loop = {'None': 0,
              'Norm': 1,
              '>200': 2,
              '>300': 3}

med_loop = {'No': 0,
            'Down': 1,
            'Steady': 2,
            'Up': 3}
```

```python
# Using my three loops to replace the data in my columns
df['A1Cresult'] = df['A1Cresult'].replace(a1_loop)

df['max_glu_serum'] = df['max_glu_serum'].replace(serum_loop)

df['metformin'] = df['metformin'].replace(med_loop)
df['repaglinide'] = df['repaglinide'].replace(med_loop)
df['nateglinide'] = df['nateglinide'].replace(med_loop)
df['glimepiride'] = df['glimepiride'].replace(med_loop)
df['glipizide'] = df['glipizide'].replace(med_loop)
df['glyburide'] = df['glyburide'].replace(med_loop)
df['pioglitazone'] = df['pioglitazone'].replace(med_loop)
df['rosiglitazone'] = df['rosiglitazone'].replace(med_loop)
df['acarbose'] = df['acarbose'].replace(med_loop)
df['insulin'] = df['insulin'].replace(med_loop)
df['glyburide-metformin'] = df['glyburide-metformin'].replace(med_loop)
```

```python
# Replacing readmitted so i can work with it. Decided to use just Yes and No to simply y variable
df.readmitted.replace({'NO': 0, #No
                       '>30': 1, #Yes
                       '<30': 1}, inplace=True)
```

```python
# Replacing diabetesMed so I can work with it
df.diabetesMed.replace({'Yes': 1,
                        'No': 0}, inplace=True)
```

```python
# Replacing change so i can work with it
df.change.replace({'Ch': 1,
                   'No': 0}, inplace=True)
```

```python
df.gender.replace({'Female': 0,
                   'Male': 1}, inplace=True)
```

```python
color_dict = {0: 'red',
              1: 'blue'}

color_series = df['readmitted'].map(color_dict)
```

```python
# Turning all of my strings into floats
def convert_strings(string):
        try:
            return float(string)
        except:
            return 1000
    
df['diag_1'] = df['diag_1'].apply(convert_strings)
df['diag_2'] = df['diag_2'].apply(convert_strings)
df['diag_3'] = df['diag_3'].apply(convert_strings)
```

```python
df['diag_1'] = df['diag_1'].astype(int)
df['diag_2'] = df['diag_2'].astype(int)
df['diag_3'] = df['diag_3'].astype(int)
```

```python
def category(num):
    if 1 <= num <= 139:
        return 1
    elif 140 <= num <= 239:
        return 2
    elif 240 <= num <= 279:
        return 3
    elif 280 <= num <= 289:
        return 4
    elif 290 <= num <= 319:
        return 5
    elif 320 <= num <= 389:
        return 6
    elif 390 <= num <= 459:
        return 7
    elif 460 <= num <= 519:
        return 8
    elif 520 <= num <= 579:
        return 9
    elif 580 <= num <= 629:
        return 10
    elif 630 <= num <= 679:
        return 11
    elif 680 <= num <= 709:
        return 12
    elif 710 <= num <= 739:
        return 13
    elif 740 <= num <= 759:
        return 14
    elif 760 <= num <= 779:
        return 15
    elif 780 <= num <= 799:
        return 16
    elif 800 <= num <= 999:
        return 17
    else:
        return 18
```
